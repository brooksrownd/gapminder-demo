```python
import pandas as pd
```

```python
url = "https://raw.githubusercontent.com/MikeTrizna/binder_demo/master/gapminder_data.tsv"
gapminder = pd.read_csv(url,sep="\t")
gapminder
```

```python
import plotly.express as px
```

```python
fig = px.scatter(gapminder,
                x='gdpPercap',
                y='lifeExp',
                size='pop',
                color='continent',
                hover_name='country',
                log_x=True,
                size_max=60)
```

```python
fig.show()
```
